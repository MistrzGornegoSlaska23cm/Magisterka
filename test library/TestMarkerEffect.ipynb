```python
import sys
import os
import importlib
import matplotlib.pyplot as plt
from PIL import Image

sys.path.insert(0, os.path.abspath('..'))
sys.path.insert(0, os.path.abspath('../Augmente'))

import MarkerEffect as marker_effect_module
importlib.reload(marker_effect_module)
RandomMarkerEffect = marker_effect_module.RandomMarkerEffect
```

```python
marker_transform = RandomMarkerEffect(
    p=1.0,
    marker_size_range=(10, 50),
    alpha_range=(90, 180),
)
```

```python
sample_img_dir = os.path.join('..', 'dataset_split', 'train', 'image')
image_files = [f for f in os.listdir(sample_img_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff'))]
sample_img = Image.open(os.path.join(sample_img_dir, image_files[0])).convert('RGB')
```

```python
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

axes[0].imshow(sample_img)
axes[0].set_title('Original Image')
axes[0].axis('off')

for i in range(1, 4):
    augmented_img = marker_transform(sample_img)
    axes[i].imshow(augmented_img)
    axes[i].set_title(f'Augmented Image {i}')
    axes[i].axis('off')
    
plt.tight_layout()
plt.show()
```
