```python
import sys
import os
import importlib

sys.path.insert(0, os.path.abspath('..'))
sys.path.insert(0, os.path.abspath('../Augmente'))
import matplotlib.pyplot as plt
from PIL import Image
import InkEffect as ink_effect_module
importlib.reload(ink_effect_module)
from InkEffect import RandomInkEffect
```

```python
ink_transform = RandomInkEffect(
    p=1.0,
    int_size_range=(22, 45)
)
```

```python
sample_img_dir = os.path.join('..', 'dataset_split', 'test', 'image')
image_files = [f for f in os.listdir(sample_img_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.gif', 
                                                                            '.webp', '.tif', '.jfif', '.pjpeg', '.pjp'))]
sample_img  = Image.open(os.path.join(sample_img_dir, image_files[0])).convert('RGB')
```

```python
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

axes[0].set_title('Original Image')
axes[0].imshow(sample_img)
axes[0].axis('off')

for i in range(1, 4):
    augmented_image = ink_transform(sample_img)
    axes[i].set_title(f'Augmented Image {i}')
    axes[i].imshow(augmented_image)
    axes[i].axis('off')
    
plt.tight_layout()
plt.show()
```
